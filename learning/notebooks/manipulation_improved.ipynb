```python


x_data, y_data, y_dataerr = [], [], []
train_x_data, train_y_data = [], []
times = [datetime.now()]


def progress(num_steps, metrics):
  clear_output(wait=True)

  times.append(datetime.now())

  # # Training episode rewards (logged continuously via log_training_metrics)
  # if "episode/reward" in metrics:
  #   train_x_data.append(num_steps)
  #   train_y_data.append(metrics["episode/reward"])

  if "episode/sum_reward" in metrics:
    train_x_data.append(num_steps)
    train_y_data.append(metrics["episode/sum_reward"])

  # Eval rewards (logged at eval intervals)
  if "eval/episode_reward" in metrics:
    x_data.append(num_steps)
    y_data.append(metrics["eval/episode_reward"])
    y_dataerr.append(metrics["eval/episode_reward_std"])

  num_ts = ppo_params["num_timesteps"] if PPO else ars_params["num_timesteps"]
  plt.xlim([0, num_ts * 1.25])
  plt.xlabel("# environment steps")
  plt.ylabel("reward per episode")

  if train_y_data:
    plt.plot(train_x_data, train_y_data, color="orange", alpha=0.7, label="training" if len(train_y_data) == 1 else None)
  if y_data:
    plt.errorbar(x_data, y_data, yerr=y_dataerr, color="blue", label="eval" if len(y_data) == 1 else None)

  title_parts = []
  if train_y_data:
    title_parts.append(f"train={train_y_data[-1]:.3f}")
  if y_data:
    title_parts.append(f"eval={y_data[-1]:.3f}")
  plt.title(" | ".join(title_parts))
  plt.legend()

  display(plt.gcf())

  # 2. Silently send the exact same data to W&B in the background
  logging_dict = {f"charts/{k}": v for k, v in metrics.items()}
#   wandb.log(logging_dict, step=num_steps)

```

```python

class FlatObsWrapper(wrapper.Wrapper):
  """Wraps a dict-obs env to extract a single obs key as a flat array.
  Needed because ARS does not support dictionary observations."""

  def __init__(self, env, obs_key='state'):
    super().__init__(env)
    self._obs_key = obs_key

  def reset(self, rng):
    state = self.env.reset(rng)
    return state.replace(obs=state.obs[self._obs_key])

  def step(self, state, action):
    # Restore dict obs for the inner env (it doesn't read state.obs, but be safe)
    state = self.env.step(state, action)
    return state.replace(obs=state.obs[self._obs_key])

  @property
  def observation_size(self):
    full_size = self.env.observation_size
    if isinstance(full_size, dict):
      entry = full_size[self._obs_key]
      # entry is a shape tuple like (52,) — return the int
      if hasattr(entry, '__len__') and len(entry) == 1:
        return entry[0]
      return entry
    return full_size


if PPO:

  ppo_training_params = dict(ppo_params)
  network_factory = ppo_networks.make_ppo_networks
  if "network_factory" in ppo_params:
    del ppo_training_params["network_factory"]
    network_factory_kwargs = dict(ppo_params.network_factory)
    # Remove asymmetric obs keys — DualUR5eBoxlift returns flat obs, not a dict
    network_factory_kwargs.pop("policy_obs_key", None)
    network_factory_kwargs.pop("value_obs_key", None)
    network_factory = functools.partial(
        ppo_networks.make_ppo_networks,
        **network_factory_kwargs
    )

  train_fn = functools.partial(
      ppo.train, **dict(ppo_training_params),
      network_factory=network_factory,
      progress_fn=progress,
      log_training_metrics=True,
      training_metrics_steps = 1000,
      seed=1,
      save_checkpoint_path=os.path.abspath(f"./checkpoints/run_{CHECKPOINT_IDX}_{env_name}_PPO"),
  )

if ARS:
  
  ars_training_params = dict(ars_params)
  network_factory = ars_networks.make_policy_network
  if "network_factory" in ars_params:
    del ars_training_params["network_factory"]
    network_factory_kwargs = dict(ars_params.network_factory)
    network_factory = functools.partial(
        ars_networks.make_policy_network,
        **network_factory_kwargs
    )

  # Wrap env to flatten dict obs for ARS (extracts 'state' key)
  ars_env = FlatObsWrapper(env, obs_key='state')

  train_fn = functools.partial(
      ars.train, **dict(ars_training_params),
      network_factory=network_factory,
      progress_fn=progress,
      # ARS has no save_checkpoint_path — saving is done manually after training
  )
```

```python
def obs_debug_full(obs_dict, env):
    """
    Exact debugger matching current _get_obs() layout.
    No guessing. No heuristics.
    """

    def debug_single(obs, name):
        idx = 0

        def take(n, label):
            nonlocal idx
            val = obs[idx:idx+n]
            print(f"{label:30s} [{idx}:{idx+n}] → {val}")
            idx += n
            return val

        nv = env._mjx_model.nv
        num_dof = env.num_dof

        print("\n" + "=" * 70)
        print(f"DEBUGGING: {name}")
        print("=" * 70)

        # =========================
        # Core observation
        # =========================

        joint_pos = take(num_dof, "joint_pos")

        ball_pose = take(7, "ball_pose")
        print(f"  ├─ ball_pos  → {ball_pose[:3]}")
        print(f"  └─ ball_quat → {ball_pose[3:]}")

        print("\n--- Velocities ---")
        qvel = take(nv, "qvel")

        joint_vel = qvel[env._joint_mask_vel]

        ball_body_id = env._mj_model.body(name="ball").id
        ball_qvel_idx = env._mj_model.body_dofadr[ball_body_id]

        ball_vel = qvel[ball_qvel_idx:ball_qvel_idx+3]
        ball_ang_vel = qvel[ball_qvel_idx+3:ball_qvel_idx+6]

        print(f"joint_vel     → {joint_vel}")
        print(f"ball_vel      → {ball_vel}")
        print(f"ball_ang_vel  → {ball_ang_vel}")

        print("\n--- Target (world frame) ---")
        target_pos = take(3, "target_pos")
        target_quat = take(4, "target_quat")

        print("\n--- Time ---")
        time = take(1, "time")

        # =========================
        # Privileged tail
        # =========================

        if idx < obs.shape[0]:
            print("\n--- Privileged ---")

            task = take(1, "task")
            success = take(1, "success")
            cost_g_ball = take(1, "cost_g_ball")
            cost_r_ball = take(1, "cost_r_ball")
            current_cost_g = take(1, "current_cost_g")
            current_cost_r = take(1, "current_cost_r")
            penalty_collision = take(1, "penalty_collision_real_time")

        print(f"\nTotal used: {idx} / {obs.shape[0]}")
        print("=" * 70 + "\n")

    # =========================
    # Run for both
    # =========================

    for k, v in obs_dict.items():
        debug_single(v, k)
```

```python
# rscope setup for visualization
from rscope import brax as rscope_utils
import jax
import jax.numpy as jp

# Example configuration (adjust as needed)
RSCOPE_ENVS = 16  # Number of parallel envs for rscope viewer
DETERMINISTIC_RSCOPE = True

# Example rscope_fn for reward visualization
def rscope_fn(full_states, obs, rew, done):
    """
    All arrays are of shape (unroll_length, rscope_envs, ...)
    full_states: dict with keys 'qpos', 'qvel', 'time', 'metrics'
    obs: nd.array or dict obs based on env configuration
    rew: nd.array rewards
    done: nd.array done flags
    """
    done_mask = jp.cumsum(done, axis=0)
    valid_rewards = rew * (done_mask == 0)
    episode_rewards = jp.sum(valid_rewards, axis=0)
    print(
        "Collected rscope rollouts with reward"
        f" {episode_rewards.mean():.3f} +- {episode_rewards.std():.3f}"
    )
```

```python
# Load evaluation environment.



_VISION = False
_RSCOPE_ENVS = RSCOPE_ENVS
_DETERMINISTIC_RSCOPE = True   
_SEED = 0

if TRAINING:
    eval_env = None
    if not _VISION:
        eval_env = registry.load(env_name, config=env_cfg)
    num_envs = 1
    if _VISION:
        num_envs = env_cfg.vision_config.render_batch_size

    policy_params_fn = lambda *args: None
    if _RSCOPE_ENVS is not None and _RSCOPE_ENVS > 0:
    # Interactive visualisation of policy checkpoints
        from rscope import brax as rscope_utils

    if not _VISION:
        rscope_env = registry.load(env_name, config=env_cfg)
        rscope_env = wrapper.wrap_for_brax_training(
            rscope_env,
            episode_length=ppo_params.episode_length,
            action_repeat=ppo_params.action_repeat,
            randomization_fn=ars_training_params.get("randomization_fn") if ARS else ppo_training_params.get("randomization_fn"),
        )
    else:
        rscope_env = env

    rscope_handle = rscope_utils.BraxRolloutSaver(
        rscope_env,
        ppo_params,
        _VISION,
        _RSCOPE_ENVS,
        _DETERMINISTIC_RSCOPE,
        jax.random.PRNGKey(_SEED),
        rscope_fn,
    )

    def policy_params_fn(current_step, make_policy, params):  # pylint: disable=unused-argument
        rscope_handle.set_make_policy(make_policy)
        rscope_handle.dump_rollout(params)
```

```python

if TRAINING:
    # For ARS, pass the flat-obs wrapped env; for PPO, pass the original env
    train_env = ars_env if ARS else env
    make_inference_fn, params, metrics = train_fn(
        environment=train_env,
        wrap_env_fn=wrapper.wrap_for_brax_training,
        policy_params_fn=policy_params_fn if PPO else None,
        eval_env=eval_env,
    )

    print(f"time to jit: {times[1] - times[0]}")
    print(f"time to train: {times[-1] - times[1]}")

    # --- Save ARS params manually (ARS has no built-in checkpoint saving) ---
    if ARS:
        import pickle
        ars_ckpt_dir = os.path.abspath(f"./checkpoints/run_{CHECKPOINT_IDX}_{env_name}_ARS")
        os.makedirs(ars_ckpt_dir, exist_ok=True)
        ars_ckpt_path = os.path.join(ars_ckpt_dir, "ars_params.pkl")
        with open(ars_ckpt_path, "wb") as f:
            pickle.dump(jax.device_get(params), f)
        print(f"ARS params saved to {ars_ckpt_path}")

    # wandb.finish()  # Finish the W&B run after training 


if INFERENCE and (not TRAINING):

    if PPO:
        # Load the latest brax PPO checkpoint automatically
        from brax.training.agents.ppo import checkpoint as ppo_checkpoint
        from brax.training.agents.ppo import networks as ppo_networks
        from brax.training import checkpoint as brax_checkpoint, networks as brax_networks
        from pathlib import Path
        import json as _json

        checkpoint_dir = Path(f'./checkpoints/run_{CHECKPOINT_IDX}_{env_name}_PPO').resolve()
        # Find the latest checkpoint (highest step number)
        ckpt_dirs = sorted(
            [d for d in checkpoint_dir.iterdir() if d.is_dir() and d.name.isdigit()],
            key=lambda d: int(d.name)
        )
        latest_ckpt = ckpt_dirs[-1]
        print(f"Loading checkpoint: {latest_ckpt.name} (step {int(latest_ckpt.name):,})")

        # Load config manually, working around brax bug with null kernel init fns
        config_path = latest_ckpt / 'ppo_network_config.json'
        loaded_dict = _json.loads(config_path.read_text())
        nf_kwargs = loaded_dict['network_factory_kwargs']
        if 'activation' in nf_kwargs:
            nf_kwargs['activation'] = brax_networks.ACTIVATION[nf_kwargs['activation']]
        for k in ('policy_network_kernel_init_fn', 'value_network_kernel_init_fn',
                  'q_network_kernel_init_fn', 'mean_kernel_init_fn'):
            if k in nf_kwargs:
                if nf_kwargs[k] is None:
                    del nf_kwargs[k]
                else:
                    nf_kwargs[k] = brax_networks.KERNEL_INITIALIZER[nf_kwargs[k]]
        nf_kwargs = {k: v for k, v in nf_kwargs.items() if v is not None}

        # Fix observation_size: convert {"state": {"shape": [52], ...}} -> {"state": 52, ...}
        obs_size_raw = loaded_dict['observation_size']
        if isinstance(obs_size_raw, dict):
            obs_size = {}
            for key, val in obs_size_raw.items():
                if isinstance(val, dict) and 'shape' in val:
                    shape = val['shape']
                    obs_size[key] = shape[0] if len(shape) == 1 else tuple(shape)
                else:
                    obs_size[key] = val
        else:
            obs_size = obs_size_raw

        normalize_observations = loaded_dict.get('normalize_observations', True)
        action_size = loaded_dict['action_size']

        # Load params
        params = ppo_checkpoint.load(str(latest_ckpt))

        # Rebuild network from config — use plain dicts, not ConfigDict
        ppo_network = ppo_networks.make_ppo_networks(
            observation_size=obs_size,
            action_size=action_size,
            preprocess_observations_fn=(
                brax_checkpoint.running_statistics.normalize
                if normalize_observations
                else lambda x, y: x
            ),
            **nf_kwargs,
        )
        make_inference_fn = ppo_networks.make_inference_fn(ppo_network)
        inference_fn = make_inference_fn(params, deterministic=True)
        jit_inference_fn = jax.jit(inference_fn)
        print("PPO policy loaded successfully.")

    if ARS:
        # Load saved ARS params
        import pickle
        from brax.training.agents.ars import networks as ars_networks
        from brax.training import checkpoint as brax_checkpoint

        ars_ckpt_path = os.path.abspath(f"./checkpoints/run_{CHECKPOINT_IDX}_{env_name}_ARS/ars_params.pkl")
        with open(ars_ckpt_path, "rb") as f:
            params = pickle.load(f)
        print(f"Loaded ARS params from {ars_ckpt_path}")

        # params is (normalizer_params, policy_params)
        normalizer_params, policy_params = params
        obs_size = policy_params.shape[0]
        action_size = policy_params.shape[1]
        print(f"  obs_size={obs_size}, action_size={action_size}, policy_params shape={policy_params.shape}")

        # Rebuild ARS network and inference fn
        normalize_observations = ars_params.get('normalize_observations', True)
        ars_network = ars_networks.make_policy_network(
            observation_size=obs_size,
            action_size=action_size,
            preprocess_observations_fn=(
                brax_checkpoint.running_statistics.normalize
                if normalize_observations
                else lambda x, y: x
            ),
        )
        make_inference_fn = ars_networks.make_inference_fn(ars_network)
        inference_fn = make_inference_fn(params)
        jit_inference_fn = jax.jit(inference_fn)
        print("ARS policy loaded successfully.")
```

```python
render_every =1
renderer = mujoco.Renderer(env._mj_model, height=480, width=640)
opt = mujoco.MjvOption()
# opt.flags[mujoco.mjtVisFlag.mjVIS_TRANSPARENT] = True
opt.flags[mujoco.mjtVisFlag.mjVIS_CONTACTPOINT] = True
cam = mujoco.MjvCamera()
cam.type = mujoco.mjtCamera.mjCAMERA_FREE
cam.lookat[:] = [0, 0, 0.4]
cam.distance = 2.0
cam.azimuth = 90
cam.elevation = -30

frames = []
for state in rollout[::render_every]:
    d = mujoco.MjData(env._mj_model)
    d.qpos, d.qvel = state.data.qpos, state.data.qvel
    d.mocap_pos, d.mocap_quat = state.data.mocap_pos, state.data.mocap_quat
    mujoco.mj_forward(env._mj_model, d)
    renderer.update_scene(d, camera=cam, scene_option=opt)

    # Draw planned EEF traces as spheres
    scn = renderer.scene
    eef_0_plan = np.array(state.info['eef_0_planned'])  # (num_steps, 7)
    eef_1_plan = np.array(state.info['eef_1_planned'])  # (num_steps, 7)

    for pts, color in [(eef_0_plan, [0, 0, 1, 0.8]),    # blue
                        (eef_1_plan, [0, 0, 1, 0.8])]:  
        for j in range(len(pts)):
            if scn.ngeom >= scn.maxgeom:
                break
            # Sphere at each planned waypoint
            g = scn.geoms[scn.ngeom]
            mujoco.mjv_initGeom(
                g,
                type=mujoco.mjtGeom.mjGEOM_SPHERE,
                size=[0.008, 0, 0],
                pos=pts[j, :3],
                mat=np.eye(3).flatten(),
                rgba=np.array(color, dtype=np.float32),
            )
            scn.ngeom += 1

            # Thin capsule connecting to next waypoint
            if j < len(pts) - 1 and scn.ngeom < scn.maxgeom:
                p0, p1 = pts[j, :3], pts[j+1, :3]
                mid = (p0 + p1) / 2.0
                diff = p1 - p0
                length = np.linalg.norm(diff)
                if length > 1e-6:
                    # Build rotation matrix: z-axis along diff
                    z = diff / length
                    # Pick a non-parallel vector for cross product
                    up = np.array([0, 0, 1]) if abs(z[2]) < 0.9 else np.array([1, 0, 0])
                    x = np.cross(up, z); x /= np.linalg.norm(x)
                    y = np.cross(z, x)
                    mat = np.column_stack([x, y, z]).flatten()

                    g = scn.geoms[scn.ngeom]
                    mujoco.mjv_initGeom(
                        g,
                        type=mujoco.mjtGeom.mjGEOM_CAPSULE,
                        size=[0.005, length / 2, 0],  # radius, half-length
                        pos=mid,
                        mat=mat,
                        rgba=np.array(color, dtype=np.float32),
                    )
                    scn.ngeom += 1

    frames.append(renderer.render())
renderer.close()

length = len(rollout)
print(length)

media.show_video(frames, fps=1.0 / env.dt / render_every)
```
